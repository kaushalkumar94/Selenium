```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time

# open browser
driver = webdriver.Chrome()

# open website
driver.get("https://www.selenium.dev/selenium/web/web-form.html")

# wait for page to load
time.sleep(3)

# Text input
driver.find_element(By.NAME, "my-text").send_keys("Kaushal Kumar")

# Password
driver.find_element(By.NAME, "my-password").send_keys("123456")

# Textarea
driver.find_element(By.NAME, "my-textarea").send_keys("Learning Selenium automation")

# Dropdown select
dropdown = Select(driver.find_element(By.NAME, "my-select"))
dropdown.select_by_index(1)

# Datalist
driver.find_element(By.NAME, "my-datalist").send_keys("New York")

# File upload
driver.find_element(By.NAME, "my-file").send_keys("C:\\Users\\Public\\test.txt")

# Checkbox
driver.find_element(By.ID, "my-check-2").click()

# Radio button
driver.find_element(By.ID, "my-radio-2").click()

# Date
driver.find_element(By.NAME, "my-date").send_keys("03/12/2026")

# Submit button
driver.find_element(By.CSS_SELECTOR, "button").click()

time.sleep(5)

driver.quit()
```
